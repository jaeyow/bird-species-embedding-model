```python
import os
import shutil

source_dir = '../kaggle_data/bird-fifty/valid'
destination_dir = '../kaggle_data/bird-fifty/valid_api'

if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file == '1.jpg':
            source_file = os.path.join(root, file)
            destination_subdir = os.path.join(destination_dir, os.path.basename(root))
            if not os.path.exists(destination_subdir):
                os.makedirs(destination_subdir)
            destination_file = os.path.join(destination_subdir, 'bird.jpg')
            shutil.copy(source_file, destination_file)
```
